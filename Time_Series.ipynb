```python
def plotar(titulo, labelx, labely, x, y, dataset):
  sns.set_palette('Accent')
  sns.set_style('darkgrid')
  ax = sns.lineplot(x=x, y=y, data=dataset)
  ax.figure.set_size_inches(12, 6)
  ax.set_title(titulo, loc='left', fontsize=18)
  ax.set_xlabel(labelx, fontsize=14)
  ax.set_ylabel(labely, fontsize=14)
  ax = ax
```

```python
plotar('Aumento das vendas da Alucar de 2017 e 2018', 'Tempo', 'Aumento', 'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Aceleração das vendas da Alucar de 2017 e 2018', 'Tempo', 'Aceleração', 'mes', 'aceleracao', alucar)
```

```python
plt.figure(figsize=(16,12))
ax = plt.subplot(3,1,1)
ax.set_title('Análise de vendas Alucar de 2027 e 2018', fontsize=18, loc='left')
ax = sns.lineplot(x='mes', y='vendas', data=alucar)
plt.subplot(3,1,2)
ax.set_title('Aumento de vendas Alucar de 2027 e 2018', fontsize=18, loc='left')
ax = sns.lineplot(x='mes', y='aumento', data=alucar)
plt.subplot(3,1,3)
ax.set_title('Análise de vendas Alucar de 2027 e 2018', fontsize=18, loc='left')
ax = sns.lineplot(x='mes', y='aceleracao', data=alucar)
ax.set_title('Aceleração de vendas Alucar de 2027 e 2018', fontsize=18, loc='left')
ax = ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo1, tiulo2, titulo3):
  plt.figure(figsize=(16,12))
  ax = plt.subplot(3,1,1)
  ax.set_title(titulo1, fontsize=18, loc='left')
  ax = sns.lineplot(x=x, y=y1, data=dataset)
  plt.subplot(3,1,2)
  ax.set_title(tiulo2, fontsize=18, loc='left')
  ax = sns.lineplot(x=x, y=y2, data=dataset)
  plt.subplot(3,1,3)
  ax = sns.lineplot(x=x, y=y3, data=dataset)
  ax.set_title(titulo3, fontsize=18, loc='left')
  ax = ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'Análise de vendas Alucar de 2027 e 2018', 'Aumento de vendas Alucar de 2027 e 2018', 'Aceleração de vendas Alucar de 2027 e 2018')
```

```python
 from pandas.plotting import autocorrelation_plot
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Gráfico de Autocorrelação', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Gráfico do Aumento', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Gráfico do Aumento', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['aceleracao'][2:])
ax = ax
```

```python
print('Quantidade de linhas e colunas', vendas_por_dia.shape)
print('Quantidade de dados nulos', vendas_por_dia.isna().sum().sum())
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia['dia'] = pd.to_datetime(vendas_por_dia['dia'])
vendas_por_dia.dtypes
```

```python
vendas_por_dia['aumento'] = vendas_por_dia['vendas'].diff()
vendas_por_dia['aceleracao'] = vendas_por_dia['aumento'].diff()
vendas_por_dia.head()
```

```python
plot_comparacao('dia', 'vendas', 'aumento', 'aceleracao', vendas_por_dia, 'Análise de vendas de Outubro e Novembro - Chocoluta')
```

**Analisando a sazonalidade**

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
```

```python
vendas_por_dia.head(7)
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday': 'Terça', 'Wednesday': 'Quarta', 'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday' : 'Sabado',
       'Sunday' : 'Domingo'}
```

```python
vendas_por_dia.head()
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
```

```python
vendas_por_dia.head(14)
```

**Agrupando os dias**

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas', 'aumento', 'aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

**Correlação das vendas diárias**

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação de vendas diárias', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação do aumento de vendas diárias', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação da aceleração de vendas diárias', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['aceleracao'][2:])
ax = ax
```

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolura['vendas'],  period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ( {
    'observacao': observacao,
    'tendencia': tendencia,
    'sazonalidade': sazonalidade,
    'ruido': ruido
})
```

```python
resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(resultado.index, 'observacao', 'tendencia', 'sazonalidade', resultado, 'Exemplo de Statsmodels')
```
